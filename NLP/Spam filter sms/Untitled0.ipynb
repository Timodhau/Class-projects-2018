```python

import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.cross_validation import train_test_split  #pip install -U scikit-learn if no model_selection or from sklearn.cross_validation
from sklearn.naive_bayes import MultinomialNB
from wordcloud import WordCloud
from sklearn.cross_validation import KFold # import KFold (cross validation)
import matplotlib.pyplot as plt

df = pd.read_csv("spam.csv", encoding="ISO-8859-1")  #encoding is to adapt with special characters

# drop useless columns
df = df.drop(["Unnamed: 2","Unnamed: 3","Unnamed: 4"], axis=1)

# rename columns
df.columns = ['labels','data']

# create binary labels
df['b_labels'] = df['labels'].map({'ham':0, 'spam':1})
Y = df['b_labels'].as_matrix()

# try multiple ways of calculating features
tfidf = TfidfVectorizer(decode_error='ignore')  #Convert a collection of raw documents to a matrix of TF-IDF features.
X = tfidf.fit_transform(df['data'])  #  fit_transform method applies to feature extraction objects as TfidfTransformer           


#count_vectorizer = CountVectorizer(decode_error='ignore')  # we are ignoring problems of decoding
#X = count_vectorizer.fit_transform(df['data'])

# split up data for test and train
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, Y, test_size=0.33)


# Create the model and train it
model = MultinomialNB()
model.fit(Xtrain, Ytrain)
print("train score: ", model.score(Xtrain, Ytrain))
print("test score: ", model.score(Xtest, Ytest))


# Visualize data with word Cloud
def visualize(label):
    words = ''
    for msg in df[df['labels'] == label]['data']:
        msg = msg.lower()
        words += msg+' '
    wordcloud = WordCloud(width=600, height=400).generate(words)
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()

#visualize('spam')

#visualize('ham')
        
```

```python
# see what is going wrong
df['predictions'] = model.predict(X)

#sneaky spam is spam which bypass the filter
sneaky_spam = df[(df['predictions'] == 0) & (df['b_labels'] == 1)]['data']
for msg in sneaky_spam:
    print(msg)
```
